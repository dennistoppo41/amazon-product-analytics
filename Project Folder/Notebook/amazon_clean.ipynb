```python
import pandas as pd

# 1. Load Data
df = pd.read_csv('/content/amazon.csv')

# 2. Clean Price Columns
df["discounted_price"] = (
    df["discounted_price"].str.replace("₹", "").str.replace(",", "").astype(float)
)
df["actual_price"] = (
    df["actual_price"].str.replace("₹", "").str.replace(",", "").astype(float)
)

# 3. Clean Rating and Rating Count
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["rating_count"] = (
    df["rating_count"].str.replace(",", "").astype(float).fillna(0)
)

# 4. Extract Primary Category
df["main_category"] = df["category"].str.split("|").str[0]
df["sub_category"] = df["category"].str.split("|").str[-1]

# 5. Calculate Discount Amount
df["discount_amount"] = df["actual_price"] - df["discounted_price"]

# Save cleaned dataset
df.to_csv("amazon_cleaned.csv", index=False)
print("Data cleaned and saved to amazon_cleaned.csv!")
```
